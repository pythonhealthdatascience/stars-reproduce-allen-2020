# Transport model charts

The code below produces the figures in the report illustrating the transport results.

```python
import pandas as pd

import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
FILE_1 = '20_positive_CW.csv'
FILE_2 = '40_positive_CW.csv'
FILE_3 = '60_positive_CW.csv'

PATIENTS = 250
P_POS_1 = 0.3
P_TRANSPORT_1 = 0.75
N_REPS = 1000

fig, ax = plt.subplots(1, 3, sharey=True, sharex=False, figsize=(12,6))

results = pd.read_csv(f'output/{FILE_1}')
results.columns = ['0', '1', '2', '3', '4']
results.boxplot(ax=ax[0],column=['1', '2', '3', '4'])

results = pd.read_csv(f'output/{FILE_2}')
results.columns = ['0', '1', '2', '3', '4']
results.boxplot(ax=ax[1], column=['1', '2', '3', '4'])

results = pd.read_csv(f'output/{FILE_3}')
results.columns = ['0', '1', '2', '3', '4']
results.boxplot(ax=ax[2], column=['1', '2', '3', '4'])

#title = f'Covid-19 ambulance transport time by seating capacity to HU ({N_REPS} model runs).'
#fig.suptitle(title)
fig.subplots_adjust(hspace=0.4)
#'P(positive) = {p_pos}. P(need transport)={p_transport}')
ax[0].set_ylabel('total driving time (mins)')
#ax[0].set_xlabel('ambulance seating capacity')
#ax[1].set_xlabel('ambulance seating capacity')
#ax[2].set_xlabel('ambulance seating capacity')
ax[0].set_title('20 COVID-19 Positive')
ax[1].set_title('40 COVID-19 Positive')
ax[2].set_title('60 COVID-19 Positive')

#ax[0].annotate('*Results compare proportion Covid19+ive and ambulance seating capacity (e.g. 2 = 2 seats.)\n**Figures do not include ambulance clean-down/turnaround time.',
#            xy=(65, 11), xycoords='figure pixels')

fig.savefig('output/prelim_model_tuned.png', dpi=300, bbox_inches='tight')
```

```python
FILE_1 = '20_positive_ILS.csv'
FILE_2 = '40_positive_ILS.csv'
FILE_3 = '60_positive_ILS.csv'

N_REPS = 1000
SCALE = 2

results_20 = pd.read_csv(f'output/{FILE_1}')
results_20.columns = ['0', '1', '2', '3', '4']
results_20_double = results_20 * SCALE
```

```python
(results_20 / 60).describe()
```

```python
(results_20_double / 60).describe()
```

```python
results_40 = pd.read_csv(f'output/{FILE_2}')
results_40.columns = ['0', '1', '2', '3', '4']
results_40_double = results_40 * SCALE
(results_40 / 60).describe()
```

```python
(results_40_double / 60).describe()
```

```python
#iqr capacity 1
iqr_1 = (results_40_double / 60).describe().loc['75%']['1'] - (results_40_double / 60).describe().loc['25%']['1']
iqr_1
```

```python
iqr_1 = (results_40_double / 60).describe().loc['75%']['1'] - (results_40_double / 60).describe().loc['25%']['1']
iqr_2 = (results_40_double / 60).describe().loc['75%']['2'] - (results_40_double / 60).describe().loc['25%']['2']
iqr_3 = (results_40_double / 60).describe().loc['75%']['3'] - (results_40_double / 60).describe().loc['25%']['3']
iqr_4 = (results_40_double / 60).describe().loc['75%']['3'] - (results_40_double / 60).describe().loc['25%']['3']
```
